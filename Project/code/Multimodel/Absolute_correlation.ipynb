## Export CSV

### Whole

```python
df = pd.read_csv("csv/Whole_absolute_export.csv")
df
```

```python
table = df.round({"Correlation":3})
table.head()
```

```python
table["asterisk"] = ""
table.head()

for i in range(len(table["p-value"])):
    if table["p-value"][i] > 0.05:
        table["asterisk"][i] = " "
    if 0.05 >= table["p-value"][i] > 0.01:
        table["asterisk"][i] = "*"
    if 0.01 >= table["p-value"][i] > 0.001:
        table["asterisk"][i] = "**"
    if table["p-value"][i] <= 0.001:
        table["asterisk"][i] = "***"
        
table.head()
```

```python
table["Correlation"] = table["Correlation"].astype(str) + table["asterisk"].astype(str)
table.head()
```

```python
table = table.pivot(index="Model", columns="Variable", values="Correlation")
table
```

```python
table.to_csv("csv/Whole_export_result.csv", index = False, header = True)
```
