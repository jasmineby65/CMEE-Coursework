## Whole region

```python
df = pd.DataFrame(columns=['Model','Variable', 'Percentage change','Absolute change'])
df
```

```python
df = pd.read_csv("csv/Whole_change.csv")
df = df.replace("Mixed layer depth", "MLD")
df = df.replace("Surface temperature", "SST")
df = df.replace("Sea ice cover", "Sea ice fraction")
df
```

```python
for i in range(len(df["Absolute change"])):
    if df["Variable"][i] == "Nitrate":
        df["Absolute change"][i] = df["Absolute change"][i]/1000
    if df["Variable"][i] == "Iron":
        df["Absolute change"][i] = df["Absolute change"][i]/1e+9
    if df["Variable"][i] == "Diatom":
        df["Absolute change"][i] = df["Absolute change"][i]/ 12 / 3.154e+7
    if df["Variable"][i] == "NPP":
        df["Absolute change"][i] = df["Absolute change"][i]/ 12 / 3.154e+7

df
```

```python
df.at[42, 'Absolute change'] = (df[-1:]["Absolute change"] * 1e+9).values
```

```python
for i in range(len(df["Absolute change"])):
    if df["Variable"][i] == "Nitrate":
        df["Absolute change"][i] = df["Absolute change"][i]*1000
    if df["Variable"][i] == "Iron":
        df["Absolute change"][i] = df["Absolute change"][i]*1e+9
    if df["Variable"][i] == "Diatom":
        df["Absolute change"][i] = df["Absolute change"][i]* 12 * 3.154e+7
    if df["Variable"][i] == "NPP":
        df["Absolute change"][i] = df["Absolute change"][i]* 12 * 3.154e+7

df
```

```python
df = df.round({"Percentage change":2, "Absolute change":2})
df
```

```python
df.to_csv("csv/Whole_change.csv", header = True, index= False)
```

### Presenting

```python
df = pd.read_csv("csv/Whole_change.csv")
df = df.rename(columns={"Percentage change": "Percentage", "Absolute change": "Absolute"})
df
```

```python
df = pd.melt(df, id_vars=["Model", "Variable"], value_vars=["Percentage", "Absolute"],
             var_name='Type', value_name='Change')
df
```

```python
df = df.pivot_table(index="Variable", columns=["Model", "Type"], values=["Change"])
df
```

```python
df.to_csv("csv/Whole_change_final.csv")
```
